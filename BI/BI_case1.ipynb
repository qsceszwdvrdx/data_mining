```python
data['class_num'] = data['Class']
```

```python
data['class_num'] = data['class_num'].replace(['Good','Bad'],[1,0])
```

```python
data['class_num'].value_counts()
```

相關係數

```python
cor=pd.DataFrame(data.corr())
cor['class_num'].mean()
```

```python
good_cor=cor[cor['class_num']>0.0046]
```

```python
good_cor_list = list(good_cor.index.values)
# good_cor_list
```

```python
len(good_cor_list)
```

```python
good_cor_data = data[good_cor_list]
```

# test & train(0.1)

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X = good_cor_data.iloc[:,:-1].values
y = good_cor_data.iloc[:,-1].values
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.1, random_state=1, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
from matplotlib.colors import ListedColormap

lr = LogisticRegression(C=1.0,random_state=1)
lr.fit(X_train,y_train)
```

```python
# y_pred = lr.predict(X_test)
print('test accracy :%.3f' % lr.score(X_test,y_test))
```

kmeans

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=2, init='k-means++', n_init=10, max_iter=300, tol=1e-04, random_state=0).fit(X_train)
```

```python
y_kmeans_pred = km.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_kmeans_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(criterion='entropy', n_estimators=10, random_state=1)
forest.fit(X_train,y_train)
```

```python
y_RFC_pred = forest.predict(X_test)
accuracy_score(y_test, y_RFC_pred)
```
