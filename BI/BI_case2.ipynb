```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
fig = sm.qqplot(data['Duration'],line='45',fit='True')
plt.show()
```

```python
plt.boxplot(data['Duration'])
plt.title("Duration Boxplot")
plt.show()
```

中位數 Ｑ1,Q3 上下限

```python
A = list(data['Duration'])
A = sorted(A)
```

```python
 Q1 = (A[249]+A[250])/2
 Q3 = (A[749]+A[750])/2
 mid = (A[499]+A[500])/2
 iqr = Q3 - Q1
 max = Q3 + 1.5*iqr
 min = Q1 - 1.5*iqr
```

```python
(data['Duration'] > max).value_counts()
```
