```
isna = train.isna()
for i in isna.columns:
  print(i + ' = ' +str(isna[i].mean()))
```

```
# gender少太多了，又不好補,刪掉gender
# 原本我猜想作曲跟作詞都是null可能是作詞作曲都是表演者自己來,結果很像不是
# composer,lyricist 缺太多,又沒規律不好補,都不適合用
# isrc 只是國際標籤而已 沒有實質意義 可以不用
train = train.drop(['composer','lyricist','gender'],axis=1)
```

```
print(isna['song_length'].value_counts())
print(isna['artist_name'].value_counts())
```

```
# 用平均值來補缺值
train['song_length'].fillna( round(train['song_length'].mean()) , inplace = True)
```

```
# 本來我是想刪掉，但test也有缺artist_name
# 如果都刪掉那最後上傳的資料就會不夠多了，所以最後還是用了。
train['artist_name'].fillna('No_name', inplace = True)
```

```
isna = train.isna()
for i in isna.columns:
  print(i + ' = ' +str(isna[i].mean()))
```

```
# 最後一個是音樂的種類(genre_ids)
# 我的想法是這樣的分類可能並不準確，每個人對於種類的認知可能不太一樣，所以不用這一欄。
train = train.drop(['genre_ids'],axis=1)
```

```
train.loc[train.bd == 0 ,'bd'] = 17
```

time

```
# 單純的看注冊時間跟到期時間，可能比較沒有意義
# 所以我多加了一個time的欄位，用來計算使用時間(註冊-到期)
import datetime
import time
```

```
# 怕做錯會蓋過去，複製新的欄位來跑
train['Start'] = train['registration_init_time']
train['End'] = train['expiration_date']
```

```
# 先轉datetime算時間差
train['Start'] = pd.to_datetime(train['Start'].astype(str),format='%Y%m%d')
train['End'] = pd.to_datetime(train['End'].astype(str),format='%Y%m%d')
```

```
train['time'] = train['End'] - train['Start']
```

```
train.head(5)
```

```
# 換回int格式
train['time'] = train['time'].astype(int) /86400000000000
```

```
train = train.drop(['Start','End','registration_init_time','expiration_date'],axis=1)
```

```
train.shape
```

```
train.head(2)
```

資料類別檢查

```
train.dtypes
```

```
# 這三個雖然是用數字表示，但是轉類別比較合理
train['city'] = train['city'].astype(str)
train['registered_via'] = train['registered_via'].astype(str)
train['language'] = train['language'].astype(str)
```

```
train.dtypes
```

```
# 執行時間
starttime = datetime.datetime.now()
train.to_csv('/content/drive/MyDrive/wsdm/train_final.csv')
endtime = datetime.datetime.now()
print((endtime - starttime).seconds)
```
