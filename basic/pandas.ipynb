    改變資料

    創造新欄位

```python
df['power'] = df['HP']+df['Attack']+df['Speed']     #用欄位名稱操作比較直觀
df.head(3)
```

```python
df['total'] = df.iloc[:,4:7].sum(axis=1)      #假設把 HP,Attack,Denfense合併
df.head(3)                                    #要注意 sum不包含最後一個 要記得(+1)
```

    移除欄位

```python
df.drop(columns=['power','total'])
df.head(3)
```

    匯出檔案 dataframe >> csv

    篩選資料

```python
df1=df.loc[(df['Type 1'] == 'Grass') & (df['HP'] > 80)]
df1
```

    把舊的序號換掉

```python
df1 = df1.reset_index()
df1
```

```python
import re
df2=df.loc[~df['Name'].str.contains('B',regex=True)]     #名字裡有B的除外
df2
```

```python
df.loc[df['Type 1'].str.contains('Fire|Grass',regex=True)]   #抓出 Type 1 = Fire/Grass
```

    忽略大小寫 >> re.I

```python
df.loc[df['Type 1'].str.contains('fire|grass',flags=re.I,regex=True)]   
```

```python
df.loc[df['Name'].str.contains('pi[a-z]*',flags=re.I,regex=True)]   #有pi的內容
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I,regex=True)]   #pi開頭
```

    改變內容

```python
# df.loc[df['Type 1'] == 'Grass','Type 1'] = 'G'
df.loc[df['Type 1'] == 'G','Type 1'] = 'Grass'
df
```

```python
df.loc[df['Type 1'] == 'Grass','Legendary'] = 'True'   #把草系都當作神獸
df
```
