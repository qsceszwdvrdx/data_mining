**1.Create a program to calculate the BMI (body mass index) of a person.
Ask the user for his height in meters and his weight in kg.**

BMI = weight / (height * height)
- Weight in Kg
- Height in Meters

Print the BMI and the classification.

**Classification**

BMI less than 18.5: **Underweight**<br>
Greater than 18.5 or less than or equal to 24.9: **Normal weight**<br>
Greater than 24.9 or less than or equal to 29.9: **Overweight**<br>
Greater than 30 : **Obesity**<br>


```python
# 我是用自己的數字來填
weight = float(input('Type your weight in Kg (ex. 70.5): '))
height = float(input('Type your height in Meters (ex. 1.70): '))
BMI = weight/(height * height)
print('Your BMI is: '+ str(round(BMI,2)))
if BMI < 18.5:
    bmi_type = 'Underweight'
elif 18.5 <= BMI < 24.9:
    bmi_type = 'Normal weight'
elif 24.9 <= BMI < 29.9:
    bmi_type = 'Overweight'
elif BMI >= 30:
    bmi_type = 'Obesity'
else :
    bmi_type = 'BMI error'
    
print('The classification of your BMI is: '+ bmi_type)
```
