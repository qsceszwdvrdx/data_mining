```python
import numpy as np
import pandas as pd
```

```python
id2_w = pd.read_csv('/content/drive/MyDrive/BDA/data/match/id2_wake.csv')
id2_s = pd.read_csv('/content/drive/MyDrive/BDA/data/match/id2_sleep.csv')
id4_w = pd.read_csv('/content/drive/MyDrive/BDA/data/match/id4_wake.csv')
id4_s = pd.read_csv('/content/drive/MyDrive/BDA/data/match/id4_sleep.csv')
```

```python
id2_w['error'] = abs(id2_w['garmin_hr'] - id2_w['record_hr'])
id2_s['error'] = abs(id2_s['garmin_hr'] - id2_s['record_hr'])
id4_w['error'] = abs(id4_w['garmin_hr'] - id4_w['record_hr'])
id4_s['error'] = abs(id4_s['garmin_hr'] - id4_s['record_hr'])
```

## id3

wake

```python
import matplotlib.pyplot as plt
plt.plot(id2_w['garmin_hr'],label='garmin')
plt.plot(id2_w['record_hr'],label='record')
plt.plot(id2_w['error'],label='error')
plt.legend(loc='best')
plt.title('id3_wake')
plt.tight_layout()
plt.show()
```

```python
import scipy.stats
print(scipy.stats.shapiro(id2_w['garmin_hr']))
print(scipy.stats.shapiro(id2_w['record_hr']))
```

```python
scipy.stats.levene(id2_w['garmin_hr'], id2_w['record_hr'], center = 'mean')
```

```python
scipy.stats.f_oneway(id2_w['garmin_hr'], id2_w['record_hr'])
```

```python
print(id2_w['garmin_hr'].mean())
print(id2_w['record_hr'].mean())
print(np.std(id2_w['garmin_hr']))
print(np.std(id2_w['record_hr']))
```
