### 平均睡覺時間

```python
print(a['總睡眠時間'].mean(),a['總睡眠時間'].max(),a['總睡眠時間'].min())
print(b['總睡眠時間'].mean(),b['總睡眠時間'].max(),b['總睡眠時間'].min())
print(c['總睡眠時間'].mean(),c['總睡眠時間'].max(),c['總睡眠時間'].min())
print(d['總睡眠時間'].mean(),d['總睡眠時間'].max(),d['總睡眠時間'].min())
```

```python
print(np.std(a['總睡眠時間'])) #標準差
print(np.std(b['總睡眠時間']))
print(np.std(c['總睡眠時間']))
print(np.std(d['總睡眠時間']))
```
