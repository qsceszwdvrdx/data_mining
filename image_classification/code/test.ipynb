```python
import zipfile
import tensorflow
import numpy as np
import pandas as pd

from tqdm import tqdm
from keras.models import load_model
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import load_img, ImageDataGenerator
from tensorflow.keras.applications.resnet50 import preprocess_input
```

```python
with zipfile.ZipFile('/content/drive/MyDrive/碩二上/多媒體/data/pokemon/image_test.zip', 'r') as zf:
  zf.extractall('/content/image_test')
zf.close()
```

```python
df = pd.read_csv('/content/drive/MyDrive/碩二上/多媒體/data/pokemon/image_test.csv')
y = df['category'].values.tolist()
```

## 64

### resnet50

```python
with zipfile.ZipFile('/content/drive/MyDrive/碩二上/多媒體/model/64/resnet50.zip', 'r') as zf:
  zf.extractall('/content/resnet50(64)')
zf.close()
```

```python
resnet50_answer = []

for i in range(5):
  model = tensorflow.keras.models.load_model('/content/resnet50(64)/resnet50_' + str(i+1) + '.h5', compile=False)  
  answer = []
  print('resnet50_'+str(i+1)+' running....')
  
  for img in range(len(df['filename'])):
    img_path = '/content/image_test/' + df['filename'][img]
    img = image.load_img(img_path, target_size=(180,180))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    answer.append(np.argmax(model.predict(x))+1)
      
  resnet50_answer.append(answer)
  print('resnet50_'+str(i+1)+' finish.')
    
print('all finish')
```

```python
acc = []
for i in range(5):
  print('resnet50_'+ str(i+1) + ' score' )
  score = round(accuracy_score(y, resnet50_answer[i]),2)
  acc.append(accuracy_score(y, resnet50_answer[i]))
  print(score)
mean = round(sum(acc)/len(acc),2)
print('mean :'+str(mean))
```

```python
for i in range(5):
  print('-------------Classification Report-------------')
  print(classification_report(y, resnet50_answer[i]))
```
