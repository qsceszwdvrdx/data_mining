```python
for times in range(3):
    res = requests.get(url)
    soup = BeautifulSoup(res.text,'html.parser')
    article = soup.select('.title a')
    paging = soup.select('.btn-group.btn-group-paging a')
    last_url = 'http://ptt.cc' + paging[1]['href']
    print(times)
    for i in article:
        print(i.text,'http://www.ptt.cc' + i['href'])
    url = last_url
```

博客來爬蟲練習

```python
url='https://www.books.com.tw/web/sys_bbotm/books/020901/?o=1&v=1&page=1'                      #自己做的

for times in range(3):
    res = requests.get(url)
    soup = BeautifulSoup(res.text,'html.parser')
    article = soup.select('.msg h4 a')
    paging = soup.select('.page a')
    last_url = paging[4]['href']
    print('No'+str(times))
    for i in article:
        print(i.text,i['href'])
    url = last_url
```

```python
for page in range(1,3):                                                                      #老師做的
    url='https://www.books.com.tw/web/sys_bbotm/books/020901/?o=1&v=1&page=x'  
    url = url.replace('x',str(page))
    res = requests.get(url)
    soup = BeautifulSoup(res.text,'html.parser')
    article = soup.select('h4 a')
    print('No'+str(times))
    for i in article:
        print(i.text)
```
