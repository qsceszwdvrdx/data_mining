```python
import requests ,json
import numpy as np
import matplotlib.pyplot as plt

url = 'https://www.dcard.tw/service/api/v2/forums/dressup/posts?popular=true'
res = requests.get(url,headers={'User-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'
                        })  #çˆ¬èŸ²æ“¬äººåŒ–
resjson = json.loads(res.text)
gender_count = {'F':0, 'M':0, 'D':0}  #è¨ˆæ•¸å™¨

for outcome in resjson:
    gender_count[outcome['gender']] = gender_count[outcome['gender']]+1
    
figure_name = ['Female','Man','robot']
total = [gender_count['F'],gender_count['M'],gender_count['D']]

plt.bar(figure_name,total)
# plt.rcParams['font.sans-serif'] = ['DFKai-SB']
```
