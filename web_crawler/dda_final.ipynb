```python
import requests,json
import pandas as pd
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
```

## 104 網站

```python
url = 'https://www.104.com.tw/jobs/search/?ro=0&kwop=11&keyword=%E6%B0%B4%E6%9E%9C&expansionType=job&order=14&asc=0&page=1&mode=s&jobsource=2018indexpoc&langFlag=0&langStatus=0'
```

```python
res = requests.get(url,headers={'User-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'})
soup = BeautifulSoup(res.text,'html.parser')  
```

```python
job_com=[]
for i in soup.select('.b-block__left a'):
  job_com.append(i.text)
```

```python
print(job_com)
```

```python
job,com=[],[]
for n in range(len(job_com)):
  if n%2 == 1:
    com.append(job_com[n])
  if n%2 == 0:
    job.append(job_com[n])
```

```python
content=[]
for i in soup.select('.b-list-inline.b-clearfix.job-list-intro.b-content li'):
  content.append(i.text)
```

```python
print(content)
```

```python
content=content[12:]
```

```python
position,exp,edu=[],[],[]
for n in range(len(content)):
  if n%3 == 0:
    position.append(content[n])
  if n%3 == 1:
    exp.append(content[n])
  if n%3 == 2:
    edu.append(content[n])
```

```python
detail=[]
for i in soup.select('.job-list-tag.b-content'):
  detail.append(i.text)
```

```python
print(detail)
```

```python
detail = detail[2:]
```

```python
money=[]
for i in detail:
  money.append(i.split()[0])
```

```python
data = pd.DataFrame({'job':job,'com':com,'position':position,'exp':exp,'edu':edu,'money':money})
```

迴圈

```python
job_com=[]
job,com=[],[]
content=[]
position,exp,edu=[],[],[]
detail=[]
money=[]
```

```python
for i in range(5):
  url = 'https://www.104.com.tw/jobs/search/?ro=0&kwop=11&keyword=%E6%B0%B4%E6%9E%9C&expansionType=job&order=14&asc=0&page='+str(i+1)+'&mode=s&jobsource=2018indexpoc&langFlag=0&langStatus=0'
  res = requests.get(url,headers={'User-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'})
  soup = BeautifulSoup(res.text,'html.parser')

  for j in soup.select('.b-block__left a'):
    job_com.append(j.text)

  for n in range(len(job_com)):
    if n%2 == 1:
      com.append(job_com[n])
    if n%2 == 0:
      job.append(job_com[n])
  job_com=[]
  
  for x in soup.select('.b-list-inline.b-clearfix.job-list-intro.b-content li'):
    content.append(x.text)
  content=content[12:]

  for n in range(len(content)):
    if n%3 == 0:
      position.append(content[n])
    if n%3 == 1:
      exp.append(content[n])
    if n%3 == 2:
      edu.append(content[n])
  content=[]

  for d in soup.select('.job-list-tag.b-content'):
    detail.append(d.text)
  detail = detail[2:]

  for m in detail:
    money.append(m.split()[0])
  detail=[]
```

```python
data = pd.DataFrame({'job':job,'com':com,'position':position,'exp':exp,'edu':edu,'money':money})
```

## 圖表

```python
print(list(data['exp'].value_counts().index))
print(list(data['exp'].value_counts().values))
```

```python
exp_type = list(data['exp'].value_counts().index)
exp_number = list(data['exp'].value_counts().values)
```

中文 >>> 英文

```python
for i in range(len(exp_type)):
  if '1年' in exp_type[i]:
    exp_type[i]='one year'
  elif '2年' in exp_type[i]:
    exp_type[i]='two year'
  elif '3年' in exp_type[i]:
    exp_type[i]='three year'
  elif '5年' in exp_type[i]:
    exp_type[i]='five year'
  elif '經歷不拘' in exp_type[i]:
    exp_type[i]='no'
  else:
    exp_type[i]='other'
```

```python
fig = plt.figure(figsize =(10, 7))
plt.pie(exp_number, labels = exp_type)
plt.show()
```

```python
print(list(data['edu'].value_counts().index))
print(list(data['edu'].value_counts().values))
```

```python
edu_type = list(data['edu'].value_counts().index)
edu_number = list(data['edu'].value_counts().values)
```

中文 >>> 英文

```python
for i in range(len(edu_type)):
  if edu_type[i]=='學歷不拘':
    edu_type[i]='no'
  elif edu_type[i]=='高中':
    edu_type[i]='high_school'
  elif edu_type[i]=='專科':
    edu_type[i]='junior_college'
  elif edu_type[i]=='高中以下':
    edu_type[i]='under high school'
  elif edu_type[i]=='大學':
    edu_type[i]='uni'
  else:
    edu_type[i]='other'
```

```python
fig = plt.figure(figsize =(10, 7))
plt.pie(edu_number, labels = edu_type)
plt.show()
```
