```python
import requests
from tkinter import *
from bs4 import BeautifulSoup
```

```python
def ptt_crawler():
    url = url_input.get()
    
    for times in range(10):
        res = requests.get(url)
        soup = BeautifulSoup(res.text,'html.parser')
        art = soup.select('.title a')
        paging = soup.select('.btn-group.btn-group-paging a')
        last_url = 'https://www.ptt.cc' + paging[1]['href']
        url = last_url
        
        for i in art:
            title = i.text
            text.insert(END,title)
            text.see(END)
            text.update()
```

```python
def main():
    global url_input,text
    background = Tk()
    background.title('ptt文章標題下載')
    background.geometry('600x600')
    Label(background,text='請輸入ptt看板：',font=('微軟雅黑',20),fg='#FF5809').grid()
    url_input=Entry(background,font=('微軟雅黑',15))
    url_input.grid(row=0,column=1)
    text = Listbox(background,font=('微軟雅黑',15),width=60,height=20)
    text.grid(row=1,columnspan=2)
    Button(background,text='下載',font=('微軟雅黑',15),command=ptt_crawler).grid(row=2,column=0,sticky=W)
    Button(background,text='離開',font=('微軟雅黑',15),command=background.destroy).grid(row=2,column=1,sticky=E)
    mainloop()
```

```python
main()
```
